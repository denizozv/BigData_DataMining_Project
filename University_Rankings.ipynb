Gender Ratio Analysis

```python
# Calculating the average gender ratio for the top 100 universities
top_100 = data.head(100)
avg_female_ratio_top_100 = top_100['Female Ratio'].mean()
avg_male_ratio_top_100 = top_100['Male Ratio'].mean()

print(f"Average Female Ratio in Top 100 Universities: {avg_female_ratio_top_100:.2f}%")
print(f"Average Male Ratio in Top 100 Universities: {avg_male_ratio_top_100:.2f}%")
```

 Hypothesis Testing

We'll perform a t-test to determine if the gender ratio in the top 100 universities is significantly different from the global average.
//En iyi 100 üniversitedeki cinsiyet oranının küresel ortalamadan önemli ölçüde farklı olup olmadığını belirlemek için bir t testi yapacağız.

T Test; It examines whether there is a mean difference between two sample groups.

```python
from scipy.stats import ttest_1samp

# Global average gender ratio
global_avg_female_ratio = data['Female Ratio'].mean()

# T-test
t_stat, p_value = ttest_1samp(top_100['Female Ratio'], global_avg_female_ratio)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("The gender ratio in the top 100 universities is significantly different from the global average.")
else:
    print("There's no significant difference in the gender ratio between the top 100 universities and the global average.")
```

Geographical Analysis

Average female ratio by country  

```python
import plotly.express as px

# Reverse one-hot encoding to get the 'Country' column
data['Country'] = data.filter(like='Location_').idxmax(axis=1).str.replace('Location_', '')

# Grouping data by country and calculating the average gender ratio
country_gender_ratio = data.groupby('Country')[['Female Ratio', 'Male Ratio']].mean().reset_index()

# Visualizing the gender ratio by country using Plotly
fig = px.bar(country_gender_ratio.sort_values(by='Female Ratio', ascending=False), 
             x='Female Ratio', 
             y='Country', 
             orientation='h', 
             title='Average Female Ratio by Country', 
             labels={'Female Ratio': 'Female Ratio (%)', 'Country': 'Country'})

fig.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv("Dataset.csv")
```

```python
df.head()
```

```python
y = df['University Rank']
x = df.drop(['University Rank'], axis = 'columns')
print(x.head())
print(y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

Model training

```python
model = RandomForestClassifier(criterion="entropy", n_estimators=200)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
